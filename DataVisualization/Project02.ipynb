### Quay lại bước khám phá dữ liệu

#### Với mỗi cột có kiểu dữ liệu dạng số, các giá trị được phân bố như thế nào? <br>
Với các cột có kiểu dữ liệu số, ta sẽ tính:
- Tỉ lệ % (từ 0 đến 100) các giá trị thiếu 
- Giá trị min
- Giá trị lower quartile (phân vị 25)
- Giá trị median (phân vị 50)
- Giá trị upper quartile (phân vị 75)
- Giá trị max

```python
def calculate_quartile(data):
    nume_col_info_df = pd.DataFrame()
    dont_touch = ['User ID', 'Playlist ID', 'Duration', 'ID']
    for col in data.keys():
        if data[col].dtype in [np.int64, np.float64] and col not in dont_touch:
            missing_percentage = round(data[col].isnull().sum() * 100 / len(data[col]), 1)
            min_value = data[col].min()
            lower_quartile = data[col].quantile(0.25)
            median = data[col].median()
            upper = data[col].quantile(0.75)
            max_value = data[col].max()
            row_line = pd.Series([missing_percentage, min_value, lower_quartile, median, upper, max_value], 
                                 index = ['missing_percentage', 'min', 'lower_quartile', 'median', 'upper_quartile', 'max'])
            nume_col_info_df[col] = row_line
    return nume_col_info_df
```

#### Tạo cột dữ liệu tính thời gian tạo của playlist (khoảng cách từ ngày tạo đến thời gian hiện tại

```python
#Thời gian hiện tại
import datetime
#Thời gian hiện tại theo múi giờ chuẩn
now = datetime.datetime.now(tz=datetime.timezone.utc)

playlist_df1['Created Time'] = (datetime.datetime.today() - playlist_df1['Created Date']).astype('timedelta64[D]')
playlist_df1['Created Time']
```

```python
playlist_df1.loc[:,['Created Time','Likes Count']].describe()
```

Vì khoảng giá trị của 2 cột là khác nhau nên ta sẽ chuẩn hóa Min Max Scaler cho dữ liệu ở 2 cột

```python
min_created = playlist_df1['Created Time'].min()
max_created = playlist_df1['Created Time'].max()
playlist_df1['Created Time'] = (playlist_df1['Created Time'] - min_created) / (max_created -min_created)
playlist_df1['Created Time']
```

`Dự đoán 1`:Một playlist có thể loại A thì liệu tất cả track trong đó đều cùng thể loại A?

```python
# Lấy tất cả thể loại playlist 
playlist_genre_uni = playlist_df['Genre'].dropna().unique()
# Số lượng thể loại playlist 
num_genre_playlist = len(playlist_genre_uni)
num_genre_playlist
```

```python
# Lấy tất cả các thể loại playlist
track_genre_uni = track_df['Genre'].dropna().unique()
# Số lượng thể loại track
num_genre_track = len(track_genre_uni)
num_genre_track
```

```python
# Tính toán để kiểm chứng giả thuyết
playlist_has_genre_df = playlist_df[playlist_df['Genre'].notna()]
dup_playlist_has_genre_df = playlist_has_genre_df.groupby('Genre')

# Lấy danh sách các tracks(có thể có duplicated tracks) của từng genre playplist 
list_genre_tracks_of_playlists = []
for i in range(num_genre_playlist):
    list_tracks_temp = []
    for tracks in dup_playlist_has_genre_df.get_group(playlist_genre_uni[i])['Tracks']:
        for track in tracks.split(';'):
            list_tracks_temp.append((track_df.loc[track_df['ID'] == track]['Genre'].values)[0])
    list_genre_tracks_of_playlists.append(list_tracks_temp)

# Tính tỉ lệ phần phần trăm các track cùng và không cùng thể loại với từng playlist 
same_genre = []
different_genre = []
for i in range(num_genre_playlist):
    same_percentage = round(list_genre_tracks_of_playlists[i].count(playlist_genre_uni[i]) / 
                     len(list_genre_tracks_of_playlists[i]) * 100, 1)
    same_genre.append(same_percentage)
    different_genre.append(100-same_percentage)
```

```python
# Tạo data_frame để vẽ sơ đồ
same_of_diff ={
    "Same Genre":same_genre,
    "Different Genre":different_genre
}

df=pd.DataFrame(same_of_diff,index=playlist_genre_uni)
df
```

**Nhận xét**: Dựa vào đồ thị, ta nhận ra dự đoán ban đầu của chúng ta là **sai**, một playlist có thể loại **A** thì không hẳn toàn bộ các track trong playlist sẽ có thể loại **A**.

`Dự đoán 2`: Thời gian chạy của playlist có đúng bằng thời gian chạy của các track cộng lại?

```python
def cal_sum_duration_all_track(playlist_id):
    sum_duration = 0
    tracks_list_str = (playlist_df.loc[playlist_df['ID'] == playlist_id]['Tracks'].values)[0]
    nums_real_track = (playlist_df.loc[playlist_df['ID'] == playlist_id]['Track Count'].values)[0]
    if pd.isna(tracks_list_str):
        return 0
    else:
        tracks_list = tracks_list_str.split(';')
        for track in tracks_list:
            sum_duration += (track_df.loc[track_df['ID'] == track]['Duration'].values)[0]
        return round(sum_duration * nums_real_track / len(tracks_list), 1)

playlist_temp = []
for playlist_id in playlist_df['ID'].values:
    playlist_temp.append(cal_sum_duration_all_track(playlist_id))

playlist_df['Duration Of All Track'] = np.array(playlist_temp)
playlist_df['Duration Of All Track'] = playlist_df['Duration Of All Track'].fillna(0)
```

**Nhận xét**: nhìn tổng quan vào độ thị, ta thấy hầu hết thời lượng của playlist gần như là bằng với tổng thời lượng của các track bên trong playlist đó (ngoại trừ một vài trường hợp khi lấy dữ liệu là lấy 5 track đầu nhưng chỉ lấy được 1-2 track nên độ chênh lệch khá lớn), vậy có lẽ là dự đoán ban đầu đã đúng.

`Dự đoán 3`: Một playlist có số lượt thích cao hoặc số lượt repost cao thì tương đương với số lượt nghe cao trung bình mỗi track trong playlist cao

```python
# Số lượng thích của một playlist so với số lượt nghe trung bình mỗi track trong playlist đó
# Hàm cal_sum_time: tính số lượt nghe trung bình của mỗi track trong 1 playlist
def cal_sum_time(playlist_id):
    sum = 0
    tracks_list_str = (playlist_df.loc[playlist_df['ID'] == playlist_id]['Tracks'].values)[0]
    if pd.isna(tracks_list_str):
        return 0
    else:
        tracks_list = tracks_list_str.split(';')
        for track in tracks_list:
            sum += (track_df.loc[track_df['ID'] == track]['Playback Count'].values)[0]
        return sum / len(tracks_list)

# Thêm cột 'Playback Count Per Track' vào dataframe playlist_df
playlist_temp = []
for playlist_id in playlist_df['ID'].values:
    playlist_temp.append(cal_sum_time(playlist_id))

playlist_df['Playback Count Per Track'] = np.array(playlist_temp)
playlist_df['Playback Count Per Track'] = playlist_df['Playback Count Per Track'].fillna(0)
```
